```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from google.colab import files

uploaded = files.upload()

print("Uploaded files:")
for name in uploaded.keys():
    print("-", name)

wfo_equity = pd.read_csv("wfo_os_equity_aug.csv")
wfo_trades = pd.read_csv("wfo_os_trades_aug.csv")
wfo_params = pd.read_csv("wfo_quarter_params_aug.csv")

full_equity = pd.read_csv("full_is_equity_aug.csv")
full_trades = pd.read_csv("full_is_trades_aug.csv")
full_params = pd.read_csv("full_is_params_aug.csv")

print("WFO equity:", wfo_equity.shape)
print("WFO trades:", wfo_trades.shape)
print("WFO params:", wfo_params.shape)

print("Full IS equity:", full_equity.shape)
print("Full IS trades:", full_trades.shape)
print("Full IS params:", full_params.shape)

display(wfo_equity.head())
display(wfo_trades.head())
display(full_equity.head())
display(full_trades.head())

# Define functions for performance metrics
START_CAPITAL = 100000

def compute_equity_metrics(equity_df, start_capital=START_CAPITAL):
    """
    Calculate equity-curve based performance metrics.

    Assumptions:
    - Equity column contains account equity over time.
    - Starting capital is 100,000.
    - Sharpe ratio is calculated using daily equity returns and annualized by sqrt(252).
    """

    df = equity_df.copy()
    df["DateTime"] = pd.to_datetime(df["DateTime"])
    df = df.sort_values("DateTime").reset_index(drop=True)

    equity = df["Equity"].astype(float)

    start_time = df["DateTime"].iloc[0]
    end_time = df["DateTime"].iloc[-1]
    ending_equity = equity.iloc[-1]

    net_profit = ending_equity - start_capital

    # Drawdown
    running_max = equity.cummax()
    drawdown = equity - running_max
    max_drawdown = abs(drawdown.min())

    # Net Profit / Max Drawdown
    np_to_maxdd = net_profit / max_drawdown if max_drawdown != 0 else np.nan

    # CAGR
    years = (end_time - start_time).days / 365.25
    cagr = (ending_equity / start_capital) ** (1 / years) - 1 if years > 0 else np.nan

    # Daily returns for Sharpe
    daily_equity = df.set_index("DateTime")["Equity"].resample("D").last().dropna()
    daily_returns = daily_equity.pct_change().dropna()

    avg_daily_return = daily_returns.mean()
    daily_vol = daily_returns.std(ddof=1)

    annualized_return = avg_daily_return * 252
    annualized_volatility = daily_vol * sqrt(252)
    sharpe = annualized_return / annualized_volatility if annualized_volatility != 0 else np.nan

    return {
        "Start Date": start_time.date(),
        "End Date": end_time.date(),
        "Ending Equity": ending_equity,
        "Net Profit": net_profit,
        "Max Drawdown": max_drawdown,
        "Net Profit / Max Drawdown": np_to_maxdd,
        "CAGR": cagr,
        "Annualized Return": annualized_return,
        "Annualized Volatility": annualized_volatility,
        "Sharpe Ratio": sharpe
    }


def compute_trade_metrics(trades_df):
    """
    Calculate trade-by-trade performance metrics.

    Assumption:
    - pnl column contains profit/loss for each completed trade.
    """

    df = trades_df.copy()
    pnl = df["pnl"].astype(float)

    winners = pnl[pnl > 0]
    losers = pnl[pnl < 0]

    total_trades = len(pnl)
    win_rate = len(winners) / total_trades if total_trades > 0 else np.nan

    avg_winner = winners.mean() if len(winners) > 0 else np.nan
    avg_loser = losers.mean() if len(losers) > 0 else np.nan

    gross_profit = winners.sum()
    gross_loss = abs(losers.sum())
    profit_factor = gross_profit / gross_loss if gross_loss != 0 else np.nan

    avg_trade_pnl = pnl.mean()

    return {
        "Total Trades": total_trades,
        "Win Rate": win_rate,
        "Average Winner": avg_winner,
        "Average Loser": avg_loser,
        "Profit Factor": profit_factor,
        "Average Trade PnL": avg_trade_pnl,
        "Gross Profit": gross_profit,
        "Gross Loss": gross_loss
    }


def combine_metrics(equity_df, trades_df):
    """
    Combine equity-based and trade-based metrics into one dictionary.
    """
    equity_metrics = compute_equity_metrics(equity_df)
    trade_metrics = compute_trade_metrics(trades_df)

    combined = {}
    combined.update(equity_metrics)
    combined.update(trade_metrics)

    return combined

# Calculate performance metrics
wfo_metrics = combine_metrics(wfo_equity, wfo_trades)
full_metrics = combine_metrics(full_equity, full_trades)

# Convert into DataFrame
summary_df = pd.DataFrame({
    "WFO / OOS": wfo_metrics,
    "Full In-Sample": full_metrics
})

display(summary_df)

# Create clean performance summary table for PPT
selected_metrics = [
    "Start Date",
    "End Date",
    "Ending Equity",
    "Net Profit",
    "Max Drawdown",
    "Net Profit / Max Drawdown",
    "CAGR",
    "Annualized Return",
    "Annualized Volatility",
    "Sharpe Ratio",
    "Total Trades",
    "Win Rate",
    "Average Winner",
    "Average Loser",
    "Profit Factor",
    "Average Trade PnL"
]

ppt_summary = summary_df.loc[selected_metrics].copy()

# Format numbers for display
def format_metric(metric, value):
    if pd.isna(value):
        return ""

    if metric in ["Start Date", "End Date"]:
        return str(value)

    if metric in ["CAGR", "Annualized Return", "Annualized Volatility", "Win Rate"]:
        return f"{value:.2%}"

    if metric in ["Sharpe Ratio", "Net Profit / Max Drawdown", "Profit Factor"]:
        return f"{value:.2f}"

    if metric in ["Total Trades"]:
        return f"{int(value):,}"

    return f"{value:,.0f}"

formatted_summary = ppt_summary.copy()

for metric in formatted_summary.index:
    for col in formatted_summary.columns:
        formatted_summary.loc[metric, col] = format_metric(metric, formatted_summary.loc[metric, col])

display(formatted_summary)

# 5. Calculate decay ratios
decay_metrics = [
    "Net Profit",
    "Max Drawdown",
    "Net Profit / Max Drawdown",
    "CAGR",
    "Annualized Return",
    "Sharpe Ratio",
    "Win Rate",
    "Profit Factor",
    "Average Trade PnL"
]

decay_rows = []

for metric in decay_metrics:
    wfo_value = wfo_metrics[metric]
    full_value = full_metrics[metric]

    if full_value == 0:
        decay = np.nan
    else:
        decay = wfo_value / full_value

    decay_rows.append({
        "Metric": metric,
        "WFO / OOS": wfo_value,
        "Full In-Sample": full_value,
        "Decay Ratio": decay
    })

decay_df = pd.DataFrame(decay_rows)

display(decay_df)

# 6. Format decay table for PPT
def format_decay_value(metric, value, col):
    if pd.isna(value):
        return ""

    if metric in ["CAGR", "Annualized Return", "Win Rate"]:
        if col == "Decay Ratio":
            return f"{value:.2f}"
        return f"{value:.2%}"

    if metric in ["Sharpe Ratio", "Net Profit / Max Drawdown", "Profit Factor", "Decay Ratio"]:
        return f"{value:.2f}"

    if metric == "Max Drawdown":
        if col == "Decay Ratio":
            return f"{value:.2f}"
        return f"{value:,.0f}"

    if col == "Decay Ratio":
        return f"{value:.2f}"

    return f"{value:,.0f}"

formatted_decay = decay_df.copy()

for i, row in formatted_decay.iterrows():
    metric = row["Metric"]
    for col in ["WFO / OOS", "Full In-Sample", "Decay Ratio"]:
        formatted_decay.loc[i, col] = format_decay_value(metric, formatted_decay.loc[i, col], col)

display(formatted_decay)

# 7. Plot equity curves
def prepare_equity_for_plot(equity_df):
    df = equity_df.copy()
    df["DateTime"] = pd.to_datetime(df["DateTime"])
    df = df.sort_values("DateTime")
    return df

wfo_plot = prepare_equity_for_plot(wfo_equity)
full_plot = prepare_equity_for_plot(full_equity)

# WFO equity curve
plt.figure(figsize=(12, 5))
plt.plot(wfo_plot["DateTime"], wfo_plot["Equity"])
plt.title("AUG Gold Futures: WFO Out-of-Sample Equity Curve")
plt.xlabel("Date")
plt.ylabel("Equity")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Full In-Sample equity curve
plt.figure(figsize=(12, 5))
plt.plot(full_plot["DateTime"], full_plot["Equity"])
plt.title("AUG Gold Futures: Full In-Sample Equity Curve")
plt.xlabel("Date")
plt.ylabel("Equity")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 8. Plot drawdown curves
def add_drawdown(equity_df):
    df = equity_df.copy()
    df["DateTime"] = pd.to_datetime(df["DateTime"])
    df = df.sort_values("DateTime")
    df["Running Max"] = df["Equity"].cummax()
    df["Drawdown"] = df["Equity"] - df["Running Max"]
    return df

wfo_dd = add_drawdown(wfo_equity)
full_dd = add_drawdown(full_equity)

# WFO drawdown
plt.figure(figsize=(12, 5))
plt.plot(wfo_dd["DateTime"], wfo_dd["Drawdown"])
plt.title("AUG Gold Futures: WFO Out-of-Sample Drawdown")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Full IS drawdown
plt.figure(figsize=(12, 5))
plt.plot(full_dd["DateTime"], full_dd["Drawdown"])
plt.title("AUG Gold Futures: Full In-Sample Drawdown")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Export results
summary_df.to_csv("aug_performance_summary_raw.csv")
formatted_summary.to_csv("aug_performance_summary_for_ppt.csv")

decay_df.to_csv("aug_decay_analysis_raw.csv", index=False)
formatted_decay.to_csv("aug_decay_analysis_for_ppt.csv", index=False)

files.download("aug_performance_summary_for_ppt.csv")
files.download("aug_decay_analysis_for_ppt.csv")
```

```python
# ============================================================
# 11. Plot normalized equity curves for comparison
# ============================================================

def prepare_normalized_equity(equity_df):
    df = equity_df.copy()
    df["DateTime"] = pd.to_datetime(df["DateTime"])
    df = df.sort_values("DateTime").reset_index(drop=True)
    df["Normalized Equity"] = df["Equity"] / df["Equity"].iloc[0] * 100
    return df

wfo_norm = prepare_normalized_equity(wfo_equity)
full_norm = prepare_normalized_equity(full_equity)

plt.figure(figsize=(12, 5))
plt.plot(wfo_norm["DateTime"], wfo_norm["Normalized Equity"], linewidth=2, label="WFO / OOS")
plt.plot(full_norm["DateTime"], full_norm["Normalized Equity"], linewidth=2, label="Full In-Sample")
plt.title("AUG Gold Futures: Normalized Equity Curve Comparison")
plt.xlabel("Date")
plt.ylabel("Normalized Equity (Start = 100)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("fig2_normalized_equity_comparison.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 13. Plot decay ratio bar chart
# ============================================================

decay_plot_metrics = [
    "Net Profit",
    "Net Profit / Max Drawdown",
    "Sharpe Ratio",
    "Win Rate",
    "Profit Factor",
    "Average Trade PnL"
]

decay_plot_df = decay_df[decay_df["Metric"].isin(decay_plot_metrics)].copy()

plt.figure(figsize=(10, 5))
plt.bar(decay_plot_df["Metric"], decay_plot_df["Decay Ratio"])
plt.axhline(y=1.0, linestyle="--", linewidth=1)
plt.title("AUG Gold Futures: Performance Decay Ratios (WFO / Full IS)")
plt.xlabel("Metric")
plt.ylabel("Decay Ratio")
plt.xticks(rotation=20, ha="right")
plt.grid(True, axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig("fig4_decay_ratio_bar_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```
